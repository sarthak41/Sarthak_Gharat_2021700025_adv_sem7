```r
# For CSV file
data <- read.csv("/content/crime_dataset_india.csv", header = TRUE, sep = ",")
```

```r
library(ggplot2)
```

```r
ggplot(data,aes(x=City)) +
   geom_bar(fill='blue') +
   labs(title="number fo crimes per city",x="City",y="Number of crimes") + theme_minimal()
```

```r
weapon_distribution <- table (data$Weapon.Used)
# Create the pie chart
pie (weapon_distribution,
labels =names (weapon_distribution),
main = "Distribution of Crimes by Weapon Used",
col = rainbow(length (weapon_distribution)))
```

```r
ggplot(data, aes(x = Victim.Age)) +
geom_histogram (binwidth=5, fill= "lightblue", color = "black") +
labs (title="Distribution of Victim Ages", x = "Victim Age", y = "Count") +
theme_minimal()
```

```r
# prompt: scatter plot for victim age vs police

ggplot(data, aes(x = Victim.Age, y = Police.Deployed)) +
  geom_point() +
  labs(x = "Victim Age", y = "Police Present", title = "Scatter Plot of Victim Age vs Police Presence")
```

```r
# : bubble plot for crime type victim age and police deployed

ggplot(data, aes(x = Victim.Age, y = Police.Deployed, size = Crime.Code,color =Crime.Description)) +
  geom_point(alpha = 0.7) +
  labs(x = "Victim Age", y = "Police Deployed", size = "Crime Type", title = "Bubble Plot of Crime Type, Victim Age, and Police Deployed") +
  theme_minimal()
```
